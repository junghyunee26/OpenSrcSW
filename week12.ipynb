```python
import csv
max_temp, min_temp, gap_temp = -999, 999, -999
max_date, min_date, gap_date = '', '', ''
f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0]
        high = float(row[-1])
        low = float(row[-2])
        if high-low>gap_temp:
            gap_temp = high-low
            gap_date = date
        if low<min_temp:
            min_temp = low
            min_date = date
        if high>max_temp:
            max_temp = high
            max_date = date
f.close()
print('일교자 가장 큰 날\n'+gap_date+'\n최저기온\n'+min_date+'\n최고기온\n'+max_date)
```

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
class station:
    def __init__(self, name, num):
        self.name = name
        self.numOfUser = num
        
class route:
    def __init__(self, name):
        self.name = name
        self.high3 = []
        self.low3 = []
        for i in range(3):
            self.high3.append(station("", -1))
            self.low3.append(station("", -1))
        
    def input(self, station):
        input_numOfUser = int(station[4].replace(',','')) + int(station[5].replace(',',''))
        if self.high3[2].numOfUser < input_numOfUser:
            self.high3[2].name = station[3]
            self.high3[2].numOfUser = input_numOfUser
        if self.high3[1].numOfUser < input_numOfUser:
            self.high3[2].name = self.high3[1].name
            self.high3[2].numOfUser = self.high3[1].numOfUser
            self.high3[1].name = station[3]
            self.high3[1].numOfUser = input_numOfUser
        if self.high3[0].numOfUser < input_numOfUser:
            self.high3[1].name = self.high3[0].name
            self.high3[1].numOfUser = self.high3[0].numOfUser
            self.high3[0].name = station[3]
            self.high3[0].numOfUser = input_numOfUser
        
        if self.low3[2].numOfUser == -1 or self.low3[2].numOfUser > input_numOfUser:
            self.low3[2].name = station[3]
            self.low3[2].numOfUser = input_numOfUser
        if self.low3[1].numOfUser == -1 or self.low3[1].numOfUser > input_numOfUser:
            self.low3[2].name = self.low3[1].name
            self.low3[2].numOfUser = self.low3[1].numOfUser
            self.low3[1].name = station[3]
            self.low3[1].numOfUser = input_numOfUser
        if self.low3[0].numOfUser == -1 or self.low3[0].numOfUser > input_numOfUser:
            self.low3[1].name = self.low3[0].name
            self.low3[1].numOfUser = self.low3[0].numOfUser
            self.low3[0].name = station[3]
            self.low3[0].numOfUser = input_numOfUser
            
routes = {}
for row in data:
    if row[1] in routes:
        routes[row[1]].input(row)
    else:
        routes[row[1]] = route(row[1])
        routes[row[1]].input(row)

route_list = list(routes.keys())
for key in route_list:
    print("<" + routes[key].name + ">")
    print("이용객이 가장 많은 역 순위")
    for i in range(3):
        str = "{0:d}. {1:s} ({2:,d})".format(i+1, routes[key].high3[i].name, routes[key].high3[i].numOfUser)
        print(str)
    print("\n이용객이 가장 적은 역 순위")
    for i in range(3):
        str = "{0:d}. {1:s} ({2:,d})".format(i+1, routes[key].low3[i].name, routes[key].low3[i].numOfUser)
        print(str)
    print("\n")
```

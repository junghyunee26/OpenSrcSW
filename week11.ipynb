```python
def calc(a, b, op):
    if(op == '+'):
        return a + b

    elif(op == '-'):
        return a - b

    elif(op == '*'):
        return a * b

    elif(op == '/'):
        return a / b

class Stack:

    stack = []

    def __init__(self):
        self.stack = []
    
    def push(self, element):
        self.stack.insert(len(self.stack), element)

    def pop(self):
        self.stack.pop()

    def top(self):
        return self.stack[len(self.stack) - 1]

    def isEmpty(self):
        if(len(self.stack) == 0):
            return True
        else:
            return False

    def size(self):
        return len(self.stack)


op = Stack()
val = Stack()

d = dict()
d['+'] = 0
d['-'] = 0
d['*'] = 1
d['/'] = 1

print('계산식 입력:')
exp = input().split()

for i in exp:

    if i in '+-*/':
        if(op.isEmpty()):
            op.push(i)
            
        # 우선순위 같을 경우
        else:
            if(d[op.top()] >= d[i]):
                while(op.isEmpty() == False):
                    if(d[op.top()] >= d[i]):
                        n2 = val.top()
                        val.pop()

                        n1 = val.top()
                        val.pop()

                        val.push(calc(n1, n2, op.top()))
                        op.pop()
                    else:
                        break

                op.push(i)

            # 우선순위 높을 경우
            elif(d[op.top()] < d[i]):
                
                op.push(i)
       
    else: 
        val.push(float(i))


while(op.isEmpty() == False):
    n2 = val.top()
    val.pop()

    n1 = val.top()
    val.pop()

    val.push(calc(n1, n2, op.top()))
    op.pop()


res = val.top()

st = ""
for i in exp:
    st += str(i)

print('계산식 출력: ' + st + '=' + str(res))
```
